```python
import pandas as pd
```

```python
suicidedata_df = pd.read_csv("data/master.csv")
suicidedata_df
```

```python
renamed_suicidedata_df = suicidedata_df.rename(columns = {' gdp_for_year ($) ':'gdp_for_year',
                                                          'gdp_per_capita ($)':'gdp_per_capita',
                                                         'suicides/100k pop':'suicidesper100pop',
                                                         'HDI for year':'hdi_for_year'})
renamed_suicidedata_df['hdi_for_year'] = pd.to_numeric(renamed_suicidedata_df['hdi_for_year'], errors='coerce')
renamed_suicidedata_df['hdi_for_year'] = renamed_suicidedata_df['hdi_for_year'].fillna(0)
renamed_suicidedata_df["gdp_for_year"] = renamed_suicidedata_df["gdp_for_year"].str.replace(',', '')
renamed_suicidedata_df
```

```python
cleaned_suicidedata = renamed_suicidedata_df.drop(columns=["country-year"])
cleaned_suicidedata
```

```python
inspector = inspect(engine)
inspector.get_table_names() 
columns = inspector.get_columns('suicidedata')
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
data = pd.read_sql_query('SELECT * FROM suicidedata',con=engine)
```

```python
results = engine.execute("SELECT country, SUM(suicides_no) AS suicides FROM suicidedata GROUP BY country ORDER BY suicides DESC")

output = {}
for result in results:
    output[result['country']] = int(result['suicides'])
```

```python
print(output)
```

```python
results = engine.execute("SELECT DISTINCT country, SUM(suicides_no) AS suicides, MAX(gdp_for_year) AS gdp_for_year FROM suicidedata WHERE year = 1985 GROUP BY country;")

output = {}
for result in results:
    output[result['country']] = {
            'suicides': int(result['suicides']),
            'gdp': result['gdp_for_year']
    }
print(output)
```

```python
results = engine.execute("SELECT year, sex, SUM(suicides_no) AS suicides FROM suicidedata GROUP BY year,sex;")
output = []
for result in results:
    yearly_result = {
        'year': result['year'],
        'sex': result['sex'],
        'suicides': int(result['suicides'])
    }
    output.append(yearly_result)
```
